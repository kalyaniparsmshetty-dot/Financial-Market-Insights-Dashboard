### CREATING COPIES OF TABLES

```python
pl1_clean=pl1.copy()
pl2_clean=pl2.copy()
r1_clean=r1.copy()
r2_clean=r2.copy()
bs_clean=bs.copy()
cf_clean=cf.copy()
price_clean=price.copy()
```

### HANDALING NULL VALUES

```python
r1_clean.drop(columns=['Credit rating'], inplace=True)
# bec credit rating column has 99% missing values, so it is better to drop it from the dataset.
```

```python
for name, df in {
    "pl1": pl1_clean,
    "pl2": pl2_clean,
    "r1": r1_clean,
    "r2": r2_clean,
    "bs": bs_clean,
    "cf": cf_clean,
    "price": price_clean
}.items():
    print("\n",name)
    print((df.isnull().mean()*100).sort_values(ascending=False).head(5))


    # it is better to drop the columns with more than 50% missing values, and for the columns with less than 50% missing values, we can impute the missing values with mean or median.
```

```python
r1_clean.columns.tolist()
```

### FINDING MISSING VALUES

```python
for df in [pl1_clean, pl2_clean, r1_clean, r2_clean, bs_clean, cf_clean, price_clean]:
    num_cols = df.select_dtypes(include='number').columns
    df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# fill the missing values in the numerical columns with median values because median is more robust to outliers than mean.
```

```python
for name, df in {
    "pl1": pl1_clean,
    "pl2": pl2_clean,
    "r1": r1_clean,
    "r2": r2_clean,
    "bs": bs_clean,
    "cf": cf_clean,
    "price": price_clean
}.items():
    print(name,df.isnull().sum().sum())
```
